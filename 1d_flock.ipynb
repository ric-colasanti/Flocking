```python
import math
import matplotlib.pyplot as plt
import random as rnd
```

```python
class Agent:
    def __init__(self):
        self.pos = 0
        
    def diffuse(self):
        self.pos+=rnd.randint(-1,1)
```

```python
def test_agent():
    test_agent = Agent()
    for _ in range(10):
        test_agent.diffuse()
        print(test_agent.pos)
```

```python
population =[Agent() for _ in range(1000)]
for i in range(1000):
    for agent in population:
        agent.diffuse()
    if i>99 and i % 50 ==0:
        bins=[0]*60
        for a in population:
            bins[math.floor((a.pos+300)/10)]+=1
        plt.plot(bins)
plt.show()
        
```

```python
from scipy import stats
k2, p = stats.normaltest(bins)
```
